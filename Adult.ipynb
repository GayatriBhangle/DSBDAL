```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 
                'marital_status', 'occupation', 'relationship', 'race', 
                'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 
                'native_country', 'income']

data = pd.read_csv('adult.csv', names=column_names)
```

```python
data.head()
```

```python
# Hsitograms (Objective: Visualize the distribution of numerical)
plt.figure(figsize=(22, 15))
data.hist(bins=20)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.pie(data['education'].value_counts(), labels=data['education'].unique(), autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
# dot plots (objective: compare two continuos varibales)
plt.figure(figsize=(12, 8))
sns.stripplot(x='age', y='hours_per_week', data=data)
plt.show()
```

```python
#bar plot: -> compare a categiriccal variable with a continuaous variable
plt.figure(figsize=(12, 8))
sns.barplot(x='workclass', y='hours_per_week', data=data, estimator =sum)
plt.xticks(rotation=45)
plt.show()
```

```python
#line charts - > show trends over a continuous variable
plt.figure(figsize=(12, 8))
sns.lineplot(x='age', y='hours_per_week', data=data)
plt.show()
```

```python
#histogram and scatter plot with box plot -> show distribution, outliers, and relationships
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.histplot(x='hours_per_week', data=data, kde=True)
plt.subplot(2, 2, 1)
sns.scatterplot(x='age', y='hours_per_week', data=data)
plt.subplot(2,2,1)
sns.boxplot(x='workclass', y='hours_per_week', data=data)
plt.xticks(rotation=45)
plt.tight_layout()
```
